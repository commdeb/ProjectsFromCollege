```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
vif = None
dataset = "./owid-covid-data.csv"
r = 0.5 # akceptowalna minimalna wartość bezwzględna korelacji do scrapowania datasetu
```

```python
covid_data = pd.read_csv(dataset, delimiter=",")
display(covid_data)
```

Na potrzeby predykcji zawężam zakres dat dla krajów do lat 2021 oraz 2022, oraz wydzielam dane nienumeryczne.
Nastęnie tworzę reprezentację one-hot dla kolumny tests_units

```python

covid_data = covid_data[covid_data["date"].str[3] != "0"]
covid_data = covid_data.reset_index()
continent_country_tests_units = covid_data.iloc[:,1:5]


one_hot_tests_units = pd.get_dummies(covid_data["tests_units"], prefix="tests_units")

covid_data = covid_data.drop("tests_units", axis=1)
covid_data = pd.concat([covid_data, one_hot_tests_units], axis=1)
covid_data = covid_data.iloc[:,5:]

display(covid_data)
display(continent_country_tests_units)
```

Informację o typach danych i ich występowaniu po oczyszczeniu.

```python
print(covid_data.info())
```

Wypełniam puste dane.

```python
covid_data = covid_data.fillna(0)
display(covid_data)
```

Analiza wpływu poszczególnych danych na ilość nowych dziennych przypadków poprzez współczynnik korelacji Pearsona.
Przedstawienie poszczególnych danych i ich wartości wspólczynnika korelacji. 

```python
matplotlib.rcParams['figure.figsize'] = [20, 20]
#Y = covid_data[1].values                                         
#X = covid_data[0].values

corr_data = covid_data.iloc[:,[0,1]]

corr_data = covid_data.corr(method = "pearson")["new_cases"]
#display(data)
for key,val in corr_data.items():
    print(f"{key: <50}{val: .2f}")

#sns.scatterplot(x='total_cases', y='new_cases', data=covid_data, sizes=(20, 200))
```

Używamy Variance Inflation Factor do odrzucenia zmiennych o zbyt podobnych korelacjach z new_cases (zbędnych zmiennych niezależnych) UWAGA dane poprzednio obliczoncych VIF powinny znajdować się w pliku vif.xlsx

```python
import os
if not os.path.isfile('./vif.xlsx'): 
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    col_to_drop = ["new_cases"]
    iter = 1
    while True:
        print(iter)
        X=covid_data.drop(col_to_drop, axis=1)
        vif = pd.DataFrame()
        vif["features"] = X.columns
        vif["vif_Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
        #pd.set_option('display.max_rows', 500)
        index = vif['vif_Factor'].idxmax()
        value = vif.loc[index, 'vif_Factor']
        if value < 10:
            break
        feature = vif.loc[index, "features"]
        col_to_drop.append(feature)
        iter = iter + 1
    vif.to_excel("vif.xlsx")
```

Odrzucenie zmiennych, które mają mały (liniowy) wpływ na dzienną wartość zakażeń według wartości współczynnik korelacji Pearsona. 
Warunek odrzucenia jest określany na początku pliku jako wartość bezwzględna współczynnika korelacji który przyjmuje wartości:
\begin{equation}
r \in [-1, 1]
\end{equation}

```python
corr_data = corr_data[lambda x: abs(x) > r]
display(corr_data)
```

Wydzielenie części wspólnej z odpowiedniej wartości współczynnika korelacji oraz czynnika VIF

```python
if vif is None:
    vif = pd.read_excel("vif.xlsx")
    
display(vif)
vif = vif[vif["features"].isin(corr_data.index.values)]
new_covid_data = covid_data.filter(items=vif["features"].values)
display(new_covid_data)
```

Wydzielenie zmiennych i wartości do modelu.

```python
X = new_covid_data
Y = covid_data["new_cases"]
```

# Tworzenie modelu regresji


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error
from sklearn.compose import TransformedTargetRegressor
from sklearn.metrics import median_absolute_error, r2_score
```

Dane są dzielone na testowe i do trenowania w stosunku 1:2000 i z randomizają ustawioną na 42

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 
```

Porówanie wartości estymowanych z rzeczywistymi w tabeli

```python
model_lin = LinearRegression(fit_intercept=True)
model_lin.fit(X_train, Y_train)
Y_pred = model_lin.predict(X_test)

Y_pred_series = pd.Series(Y_pred,index=Y_test.index, name="new_cases_pred")
# display(Y_pred)

compare_results = pd.concat([Y_test, Y_pred_series], axis=1)
display(compare_results)
print(f"test sample size:     {len(Y_test)}")
print(f"training sample size: {len(Y_train)}")
```

Tabela wartości parametru zmiennych w modelu. 

```python

coef_series = pd.Series(model_lin.coef_, index=new_covid_data.columns.values)
data_with_coef = pd.DataFrame({"Coefficient":coef_series})
display(data_with_coef)
print(f"Wyraz wolny: {model_lin.intercept_}")
```

Wykres porównujący wartości estymowane z rzeczywistymi

```python
idx = list(range(int(len(Y_test) / 300),int(len(Y_test) / 200)))
Y_test_to_print = Y_test.values[idx]
Y_pred_to_print = Y_pred[idx]
X_axis = list(range(len(Y_test_to_print)))

print(Y_test_to_print[10])
print(Y_pred_to_print[10])

sns.scatterplot(x=X_axis, y=Y_test_to_print, color="orange", sizes=(20, 200))
sns.scatterplot(x=X_axis, y=Y_pred_to_print, color="blue", sizes=(20, 200))


# print(len(Y_test_to_print))
# print(len(Y_pred_to_print))
```

Wartości średniego błędu kwadratowego dla modelu oraz współczynnik determinacji R^2

```python

sq_err = mean_squared_error(Y_test, Y_pred)
print(f"mean_sq_err  = {sq_err}")
print(f"R^2 to model = {model_lin.score(X_train, Y_train)}")
print(f"R^2 to data  = {r2_score(Y_test,Y_pred)}")
print(f"med_abs_err  = {median_absolute_error(Y_test, Y_pred)}")
```
